## 1. Plotting

```python
# read image to array
img = array(Image.open('img/alcohol1.jpg').convert('L'))
```

```python
x = [100,100,400,400]
y = [200,500,200,500]

subplot(1,2,1)
imshow(img)

plot(x, y, 'r*')  # red star
plot(x[:2], y[:2])  # line connecting two points (default : blue)

title("Plotting : alcohols")

subplot(1,2,2)
imshow(img)

plot(x, y, 'r*')  # red star
plot(x[:2], y[:2])  # line connecting two points (default : blue)

axis('off')

title("Plotting : alcohols with no axis")

show()
```

```python
# visualization of image contours
figure()  # create a new fig
gray()  # don't use color

contour(img, origin = "image")  # need single value -> grayscale
axis('equal')
axis('off')
```

```python
# plotting image histograms (distribution of pixel values) - how many pixels have values in the bin's range
figure()
hist(img.flatten(), 128)  # 1st arg : 1d, 2nd arg : the number of bins to use
show()
```
