## 2. Convert Images to JPG

```python
# return all images in a dir
def get_imlist(path) : 
    return [os.path.join(path,f) for f in os.listdir(path)]
```

```python
filelist = get_imlist('img')
print(filelist)
```

```python
# Convert Images to JPG
for infile in filelist : 
    outfile = os.path.splitext(infile)[0] + ".jpg"
    if infile != outfile : 
        try : 
            Image.open(infile).save(outfile)
        except IOError : 
            print ("cannot convert", infile)  # already jpg
```

```python
# return all jpg images in a dir
def get_jpg_imlist(path) : 
    return [os.path.join(path,f) for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
jpg_filelist = get_jpg_imlist('img')
print(jpg_filelist)
```

## 4. Copy and Paste

```python
pil_im = Image.open('img/alcohol1.jpg')

# cropping a region
box = (100, 100, 400, 400)  # upper-left corner : (0,0)
region = pil_im.crop(box)  # 4-tuple - (left, upper, right, lower)
```

```python
region = region.transpose(Image.ROTATE_180)
pil_im.paste(region, box)
```

```python
display(pil_im)
```

```python
pil_im_r = pil_im.resize((128, 128))
pil_im_r = pil_im_r.rotate(45)
```
